<h1 style="color:violet;">Analyse Past Winners of the Nobel Prize</h1>

<p style="font-size:20px;">On November 27, 1895, Alfred Nobel signed his last will in Paris. When it was opened after his death, the will caused a lot of controversy, as Nobel had left much of his wealth for the establishment of a prize. Alfred Nobel dictates that his entire remaining estate should be used to endow “prizes to those who, during the preceding year, have conferred the greatest benefit to humankind”. Every year the Nobel Prize is given to scientists and scholars in the categories chemistry, literature, physics, physiology or medicine, economics, and peace</p>

```python
award_per_year = df.groupby("year").count().prize
```

```python
rolling_df = award_per_year.rolling(window=5).mean()
```

```python
plt.figure(figsize=(14, 6), dpi=120)
plt.scatter(x=award_per_year.index,
            y=award_per_year.values,
            c='dodgerblue',
            alpha=0.7,
            s=100)
plt.plot(award_per_year.index,
         rolling_df.values,
         c="crimson",
         linewidth=3)
plt.title("Number of Prize Awarded Per Year", fontsize=18)
plt.xticks(np.arange(1900, 2021, step=5), fontsize=14, rotation=45)
plt.yticks(fontsize=14);
```

```python
yearly_avg_share = df.groupby('year').agg({'percent_share':pd.Series.mean})
yearly_moving_avg = yearly_avg_share.rolling(window=5).mean()
```

```python
plt.figure(figsize=(14, 6), dpi=120)
ax1 = plt.gca()
ax2 = ax1.twinx()
ax1.scatter(x=award_per_year.index,
            y=award_per_year.values,
            c='dodgerblue',
            alpha=0.7,
            s=100)
ax1.plot(award_per_year.index,
         rolling_df.values,
         c="crimson",
         linewidth=3)

ax2.plot(award_per_year.index,
         yearly_moving_avg,
         c='gray',
         linewidth=3)

plt.title("Number of Prize Awarded Per Year", fontsize=18)
plt.xticks(np.arange(1900, 2021, step=5), fontsize=14, rotation=45)
plt.yticks(fontsize=14);
```

```python
birth_years = df.birth_date.dt.year
```

```python
df['winning_age'] = df.year - birth_years
```

```python
display(df.nlargest(n=1, columns='winning_age'))
display(df.nsmallest(n=1, columns='winning_age'))
```

```python
plt.figure(figsize=(8, 4), dpi=200)
sns.histplot(data=df,
             x=df.winning_age,
             bins=30)
plt.xlabel('Age')
plt.title('Distribution of Age on Receipt of Prize')
plt.show()
```

```python
plt.figure(figsize=(8,4), dpi=200)
with sns.axes_style("whitegrid"):
    sns.regplot(data=df,
                x='year',
                y='winning_age',
                lowess=True, 
                scatter_kws = {'alpha': 0.4},
                line_kws={'color': 'black'})
 
plt.show()
```

```python
plt.figure(figsize=(8,4), dpi=200)
with sns.axes_style("whitegrid"):
    sns.boxplot(data=df,
                x='category',
                y='winning_age')
 
plt.show()
```

```python
with sns.axes_style('whitegrid'):
    sns.lmplot(data=df,
               x='year', 
               y='winning_age',
               row = 'category',
               lowess=True, 
               aspect=2,
               scatter_kws = {'alpha': 0.6},
               line_kws = {'color': 'black'},)
 
plt.show()
```

```python
with sns.axes_style("whitegrid"):
    sns.lmplot(data=df,
               x='year',
               y='winning_age',
               hue='category',
               lowess=True, 
               aspect=2,
               scatter_kws={'alpha': 0.5},
               line_kws={'linewidth': 5})
 
plt.show()
```
